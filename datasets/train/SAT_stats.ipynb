```python
import json
from tqdm import tqdm

DATASET_PATH = "/jet/home/ydinga/idl_project/shared/datasets/SAT/SAT_data.jsonl"

train_data = []
with open(DATASET_PATH, "r") as f:
    train_data = json.load(f)

filtered_train_data = []

for example in train_data:
    if not example["correct_answer"].isdigit():
        filtered_train_data.append(example)

print("kept: " + str(len(filtered_train_data)))
print("discarded: " + str(len(train_data) - len(filtered_train_data)))

with open("filtered_train_data.jsonl", "w") as f:
    for example in filtered_train_data:
        f.write(json.dumps(example) + "\n")
```

```python
import json
import random

data = []

with open("filtered_train_data.jsonl", "r") as f:
    for line in f:
        example = json.loads(line.strip())
        data.append(example)

print(len(data))
```

```python
keywords = {
    "RELATIVE_LOCATION": ["left", "right", "above", "below", "where is", "further away", "in front of", "behind"],
    "DISTANCE_COMPARISON": ["closer to", "centerpoint", ]
}

categories = [
    "RELATIVE_LOCATION",
    "DISTANCE_COMPARISON"
]

counter = {
    "RELATIVE_LOCATION": 0,
    "DISTANCE_COMPARISON": 0,
    "OTHER": 0
}

for example in data:
    if any(keyword in example["question"].lower() for keyword in keywords["RELATIVE_LOCATION"]):
        example["category"] = "RELATIVE_LOCATION"
        counter["RELATIVE_LOCATION"] += 1
    elif any(keyword in example["question"] for keyword in keywords["DISTANCE_COMPARISON"]):
        example["category"] = "DISTANCE_COMPARISON"
        counter["DISTANCE_COMPARISON"] += 1
    else:
        example["category"] = "OTHER"
        counter["OTHER"] += 1


print(counter)

with open("data_labeled.jsonl", "w") as f:
    for example in data:
        f.write(json.dumps(example) + "\n")
```
