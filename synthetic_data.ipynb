```python
import json
import numpy as np
```

```python
category = ['height', 'orientation', 'multi-objects', 'location']
```

```python
fixed_camera = {"position": [0, 0, 0], "orientation": [0, 0, 0]}
```

```python
color_dict = [
    ['blue',   [0, 0, 1]],
    # ['red',    [1, 0, 0]],
    ['green',  [0, 1, 0]],
    ['yellow', [1, 1, 0]],
    ['purple', [1, 0, 1]],
    ['orange', [1, 0.5, 0]],
    ['brown',  [0.5, 0.25, 0]],
    ['gray',   [0.5, 0.5, 0.5]],
]
```

```python
def height(file, camera, color_dict, num_object = 2, xy_range_low = -20, xy_range_low_high = 20, min_diff = 0.5):
    json_ori = {}
    json_ques = {}

    with open(file, 'r') as f:
        json_ori = json.load(f)

    json_ques['camera'] = camera

    low_coor = {}
    low_pos = []
    low_pos.append(np.random.uniform(xy_range_low, xy_range_low_high))
    low_y = np.random.uniform(xy_range_low, xy_range_low_high)
    low_pos.append(low_y)
    low_pos.append(np.random.uniform(xy_range_low, 0))
    low_coor['position'] = low_pos
    low_coor['orientation'] = np.random.uniform(xy_range_low, xy_range_low_high, 3).tolist()

    json_ques['object_1'] = low_coor

    high_coor = {}
    high_pos = []
    high_pos.append(np.random.uniform(xy_range_low, xy_range_low_high))
    high_pos.append(np.random.uniform(low_y + min_diff, xy_range_low_high))
    high_pos.append(np.random.uniform(xy_range_low, 0))
    high_coor['position'] = high_pos
    high_coor['orientation'] = np.random.uniform(xy_range_low, xy_range_low_high, 3).tolist()


    json_ques['object_2'] = high_coor

    for obj in range(2, num_object):
        obj_coor = {}

        pos = []

        pos.append(np.random.uniform(xy_range_low, xy_range_low_high))
        pos.append(np.random.uniform(xy_range_low, xy_range_low_high))
        pos.append(np.random.uniform(xy_range_low, 0))

        obj_coor['position'] = pos
        obj_coor['orientation'] = np.random.uniform(xy_range_low, xy_range_low_high, 3).tolist()

        json_ques[f'object_{obj}'] = obj_coor

    json_ori.append(json_ques)

    with open(file, 'w') as f:
        json.dump(json_ori, f, indent = 2)
    
```
